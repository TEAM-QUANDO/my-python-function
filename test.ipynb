```python
import io
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("C:\\External\\jack0\\Downloads\\group-image-2-1625645591307")
img_t = cv2.imread("C:\\External\\jack0\\Downloads\\group-image-2-1625645591307-t")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_t = cv2.cvtColor(img_t, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title(f"Original Size {img.shape[:-1]}", fontdict={"fontsize": 25}, pad=25)
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_t)
plt.title(f"Thumbnail Size {img_t.shape[:-1]}", fontdict={"fontsize": 25}, pad=25)
plt.axis("off")

plt.show()
```
